Chart Generation Code
---------------------

This is the code used to produce the figures in the paper.

```python
#Import libraries and data.
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
import seaborn as sns
import pandas as pd

#Import data, which is generated by copy-pasting from excel sheets into .txt files.
years_for_charts = np.loadtxt('years_for_charts.txt')
year_counts = np.loadtxt('year_counts.txt')
salience_data=np.loadtxt('salience_data_level1.txt')
salience_data_L2=np.loadtxt('salience_data_level2.txt')
```

Code below creates chart showing growth in total formal epistemology abstracts over time. 

```python
#Put the data into a dataframe

df=pd.DataFrame({'xvalues': years_for_charts, 'yvalues': year_counts})
sns.set_theme()    
    
# plot
plt.plot('xvalues', 'yvalues', data=df)
plt.xlabel('Year')
plt.ylabel('Abstracts')

# show the graph
plt.savefig('overall_diachronic.png')
plt.show()
```

```python
fig, axes = plt.subplots(4,4)
df1=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,1]})
df2=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,2]})
df3=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,3]})
df4=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,4]})
df5=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,5]})
df6=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,6]})
df7=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,7]})
df8=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,8]})
df9=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,9]})
df10=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,10]})
df11=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,11]})
df12=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,12]})
df13=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,13]})
df14=pd.DataFrame({'xvalues': salience_data[:,0], 'yvalues': salience_data[:,14]})
axes[0,0].plot('xvalues', 'yvalues', data=df1)
axes[0,0].set_title('Evidence')
axes[0,0].set_ylim([0,.18])
axes[0,1].plot('xvalues', 'yvalues', data=df2)
axes[0,1].set_title('Logical Approaches to Belief Revision')
axes[0,1].set_ylim([0,.18])
axes[0,2].plot('xvalues', 'yvalues', data=df3)
axes[0,2].set_title('Logical Approaches to Knowledge')
axes[0,2].set_ylim([0,.18])
axes[0,3].plot('xvalues', 'yvalues', data=df4)
axes[0,3].set_title('Probability and Rationality')
axes[0,3].set_ylim([0,.18])
axes[1,0].plot('xvalues', 'yvalues', data=df5)
axes[1,0].set_title('Inductive Risk')
axes[1,0].set_ylim([0,.18])
axes[1,1].plot('xvalues', 'yvalues', data=df6)
axes[1,1].set_title('Applications to General PhilSci')
axes[1,1].set_ylim([0,.18])
axes[1,2].plot('xvalues', 'yvalues', data=df7)
axes[1,2].set_title('Abstracts not in English')
axes[1,2].set_ylim([0,.18])
axes[1,3].plot('xvalues', 'yvalues', data=df8)
axes[1,3].set_title('Judgement')
axes[1,3].set_ylim([0,.18])
axes[2,0].plot('xvalues', 'yvalues', data=df9)
axes[2,0].set_title('Coherence')
axes[2,0].set_ylim([0,.18])
axes[2,1].plot('xvalues', 'yvalues', data=df10)
axes[2,1].set_title('Applications to Scientific Modelling')
axes[2,1].set_ylim([0,.18])
axes[2,2].plot('xvalues', 'yvalues', data=df11)
axes[2,2].set_title('Hypothesis Testing and Confirmation')
axes[2,2].set_ylim([0,.18])
axes[2,3].plot('xvalues', 'yvalues', data=df12)
axes[2,3].set_title('Data Science Applications')
axes[2,3].set_ylim([0,.18])
axes[3,1].plot('xvalues', 'yvalues', data=df13)
axes[3,1].set_title('Applications to Scientific Practice')
axes[3,1].set_ylim([0,.18])
axes[3,2].plot('xvalues', 'yvalues', data=df14)
axes[3,2].set_title('Causal Modelling')
axes[3,2].set_ylim([0,.18])
fig.set_size_inches(18,18)
plt.savefig('salience_level1.png')
plt.show()
```
